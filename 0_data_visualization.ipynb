#### Bar Chart

```python
data_fig = pd.DataFrame({'type':['Bored' , 'Not great', 'Ok', 'Kind of interested', 'Excited'],
                         'Percent_pre': [ 0.11, 0.05,  0.4,  0.25, 0.19],
                         'Percent_post': [ 0.12, 0.06, 0.14 , 0.30,  0.38 ] })
data_fig = data_fig.set_index('type')
data_fig
```

```python
import matplotlib.pyplot as plt

def plot_figure(data):
    
    _, ax = plt.subplots(figsize = (8, 4))
    
    columns = list(data.columns)

    xticks = np.arange(0, data.shape[0])
    w = 0.5
    for col in columns:
        
        values = data[col].values
        colorpre = [(0,0, 0, 0.5) if x>=0.4 else (0,0, 0, 0.3) for x in values ]
        colorpost = [(0,0.2, 0.56, 0.6) if x>=0.3 else (0, 0.2, 0.56, 0.3) for x in values ]
        colors = colorpre if re.search('pre' ,col, re.I) else colorpost
        xticks_new = xticks - (w - 0.1)/2 if re.search('pre' ,col, re.I) else xticks + (w - 0.1)/2
        
        ax.bar(xticks_new, values, w-0.1, color = colors)
         
        for val_col, xt in zip(values, xticks_new):
            ax.text(xt, val_col-0.029,  str(int(val_col*100)) + '%', fontsize=10 , 
                color = 'white', 
                horizontalalignment='center' )

    ax.set_ylim((0, 0.6))
    ax.set_xticks(np.arange(0, 5))
    
    ax.set_xticklabels(['Bored', 'Not great', 'Ok', 'Kind of \n interested', 'Excited'],
                      fontdict ={'fontsize':12})

    ax.set_title('How do you feel about data science?', 
                 fontdict = {'fontsize':15}, 
                loc = 'left',
                pad = 11)

    ax.text(-0.5, 0.48, 
            'BEFORE program, the\nmajority of children felt\njust OK about science', 
            fontdict ={'fontsize':11})

    ax.text( 4.60, 0.06, 
            'AFTER\nprogram,\nmore children\nwere Kind of\ninterested &\nExcited about\nscience.\n', 
            fontdict ={'fontsize':11, 'color': (0,0.2, 0.56, 0.6) })

    ax.set_facecolor('white')
    ax.tick_params( left=False, bottom=False, labelleft=False)
    spines = ['left', 'right', 'top', 'bottom']
    for spine in spines:
            ax.spines[spine].set_color('white')
    
    return ax
```

```python
plot_figure(data_fig)
plt.show()
```

```python
def subplots():    
    
    fig, ax = plt.subplots(figsize = (10, 5))
    ax.tick_params(axis = 'both', direction = 'in', length = 5)
    ax.tick_params(which='minor', bottom = False, left=False)
    
    for spine in ['right', 'top']:
        ax.spines[spine].set_color('none')
        
    for spine in ['left', 'bottom']:
        ax.spines[spine].set_color('k')    
        ax.spines[spine].set_linewidth(0.5)
        
    return fig, ax
```

###  Replicate a Professional Figure

Our goal in this part if the notebook is replicate this figure from **New York Times**
"How Flight Cancellations Rippled Around the World"
https://x.com/nytimes/status/1814472846178955282

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import colormaps
```

```python
# Airports selected
airports = [
        "Bengaluru",
        "Kempegowda"
        "Melbourne", 
        "Berlin Brandenburg",
        "Zurich",
        "Istanbul",
        "London City",
        "Amsterdam Schiphol",
        "Chicago O'Hare",
        "Atlanta",
        "La Guardia",
        "Richmond",
        "Detroit",
        "Houston",
        "CancÃºn",
        "Dulles"]

np.random.seed(1234)
# Simulated data x ~ WN(0, 1) is a White Noise Process added a constant
wn = (np.random.randn(100, 15).cumsum(axis=0) + 50)

# In order to spread each line to another is added the steps (step=5) 
steps = np.cumsum(np.ones((1, 15))*5)

data_simulation = wn + steps

# Create a data frame
data_frame = pd.DataFrame(data_simulation, columns=airports)

# Get the Viridis colormap
viridis = plt.colormaps['viridis']

# Get discrete colors based on a map viridis
colors = viridis(np.linspace(0, 1, 15))

# Order in which the column must be plotted
columns = data_frame.min(axis=0).sort_values(ascending =False).index

# Create the fig and ax
fig, ax = plt.subplots(figsize=(8, 8))


ax.tick_params(axis='both', left = False, bottom = False, labelbottom =False, labelleft = False)
spines = ['bottom', 'top', 'right', 'left']

for spine in spines:
    if spine == 'left':
        ax.spines[spine].set_color((0.1, 0.1, 0.1))
    else:
        ax.spines[spine].set_color('none')

fontdict_base={'size':9, 'ha':'right','color':(0.1, 0.1, 0.1)}
fontdict_bold = fontdict_base.copy()
fontdict_bold['weight'] = 'bold'
fontdict_bold['size'] = 10

y_text = 0
for index, (col, color) in enumerate(zip(columns, colors)):
    value_min = np.min(data_frame[col])
    idx = data_frame.index.values
    base = np.ones_like(idx)
    
    ax.fill_between(idx, base * value_min, data_frame[col], color = color)
    ax.plot(data_frame[col], color = '#FFFFFF', lw=0.18)
    
    y = value_min - 2 if abs(y_text-value_min)<2 else value_min
    if index == 0:
        ax.text(-8, value_min, col,  fontdict=fontdict_base)
    else:
        ax.text(-8, y, col,  fontdict=fontdict_base)
    y_text = y

ax.text(-8, np.max(data_frame), "AIRPORT", fontdict=fontdict_bold);
```
